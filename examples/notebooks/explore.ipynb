## 3. Change-Point Model Engine

The change-point model engine performs statistical energy modeling to identify heating/cooling relationships.

````python
from better_lbnl_os.core.changepoint import fit_changepoint_model
from better_lbnl_os.models import ChangePointModelResult

# Generate synthetic energy consumption data correlated with temperature
# This simulates a building with heating and cooling loads

# Monthly temperature data (Celsius) for a full year
monthly_temps = [7, 9, 12, 16, 20, 25, 28, 27, 23, 18, 12, 8]  # Jan-Dec

# Synthetic monthly energy consumption (kWh/day)
# Base load + heating when cold + cooling when hot
base_load = 1000  # kWh/day
heating_energy = [max(0, (18 - temp) * 150) for temp in monthly_temps]  # Heat when < 18Â°C
cooling_energy = [max(0, (temp - 22) * 200) for temp in monthly_temps]  # Cool when > 22Â°C

monthly_energy = [base_load + heat + cool for heat, cool in zip(heating_energy, cooling_energy)]

print("ğŸ“Š Synthetic Building Energy Data:")
for i, (temp, energy) in enumerate(zip(monthly_temps, monthly_energy), 1):
    print(f"  Month {i:2d}: {temp:4.1f}Â°C â†’ {energy:5.0f} kWh/day")

print(f"\nTemperature range: {min(monthly_temps):.1f}Â°C to {max(monthly_temps):.1f}Â°C")
print(f"Energy range: {min(monthly_energy):.0f} to {max(monthly_energy):.0f} kWh/day")
````

````python
# Fit change-point model to the data using new generic interface
import numpy as np

temperature_array = np.array(monthly_temps)
energy_array = np.array(monthly_energy)

# Use the new generic x,y interface - much cleaner!
model_result = fit_changepoint_model(
    x=temperature_array,  # Generic x parameter
    y=energy_array        # Generic y parameter
)

print("ğŸ¯ Change-Point Model Results (Generic Interface):")
print(f"  Model Type: {model_result.model_type}")
print(f"  RÂ²: {model_result.r_squared:.3f}")
print(f"  CV-RMSE: {model_result.cvrmse:.3f}")
print(f"  Model Valid: {model_result.is_valid()}")

if model_result.baseload is not None:
    print(f"  Base Load: {model_result.baseload:.1f} kWh/day")

if model_result.heating_slope is not None:
    print(f"  Heating Slope: {model_result.heating_slope:.1f} kWh/day/Â°C")
    if model_result.heating_change_point is not None:
        print(f"  Heating Change Point: {model_result.heating_change_point:.1f}Â°C")

if model_result.cooling_slope is not None:
    print(f"  Cooling Slope: {model_result.cooling_slope:.1f} kWh/day/Â°C") 
    if model_result.cooling_change_point is not None:
        print(f"  Cooling Change Point: {model_result.cooling_change_point:.1f}Â°C")

print(f"\nğŸŒŸ Generic Interface Benefits:")
print(f"  âœ… Works for any x,y relationship (temperature-energy, price-demand, etc.)")
print(f"  âœ… Simpler, more intuitive parameter names")
print(f"  âœ… Domain-specific result model keeps meaningful names for energy analysis")

# Example: Could also be used for other domains
print(f"\nğŸ’¡ Other possible applications:")
print(f"  â€¢ Price vs Demand analysis: fit_changepoint_model(prices, demand)")
print(f"  â€¢ Time series analysis: fit_changepoint_model(time, usage)")  
print(f"  â€¢ Economic modeling: fit_changepoint_model(income, spending)")
````

## 5. Statistical Utilities

The statistics utilities provide common statistical calculations for energy analysis.

````python
from better_lbnl_os.utils.statistics import calculate_r_squared, calculate_cvrmse

# Statistical analysis of our change-point model fit
predicted_energy = []

# Use the model to predict energy for each temperature point
for temp in monthly_temps:
    prediction = model_result.baseload or 0
    
    if model_result.heating_slope and model_result.heating_change_point:
        if temp < model_result.heating_change_point:
            prediction += model_result.heating_slope * (model_result.heating_change_point - temp)
    
    if model_result.cooling_slope and model_result.cooling_change_point:
        if temp > model_result.cooling_change_point:
            prediction += model_result.cooling_slope * (temp - model_result.cooling_change_point)
    
    predicted_energy.append(prediction)

# Calculate statistical metrics manually to verify model results
manual_r_squared = calculate_r_squared(np.array(monthly_energy), np.array(predicted_energy))
manual_cvrmse = calculate_cvrmse(np.array(monthly_energy), np.array(predicted_energy))

print("ğŸ“ˆ Statistical Analysis:")
print(f"  Model RÂ²: {model_result.r_squared:.3f}")
print(f"  Manual RÂ²: {manual_r_squared:.3f}")
print(f"  Model CV-RMSE: {model_result.cvrmse:.3f}")
print(f"  Manual CV-RMSE: {manual_cvrmse:.3f}")

print(f"\nğŸ¯ Model Performance:")
if model_result.r_squared > 0.8:
    print("  âœ… Excellent fit (RÂ² > 0.8)")
elif model_result.r_squared > 0.6:
    print("  âœ… Good fit (RÂ² > 0.6)")
else:
    print("  âš ï¸ Poor fit (RÂ² < 0.6)")

if model_result.cvrmse < 0.2:
    print("  âœ… Low prediction error (CV-RMSE < 20%)")
elif model_result.cvrmse < 0.5:
    print("  âœ… Acceptable error (CV-RMSE < 50%)")
else:
    print("  âš ï¸ High prediction error (CV-RMSE > 50%)")
````
